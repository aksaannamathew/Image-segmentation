```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("MCF7_Phase_E4_1_00d00h00m_3.tif", 0)
```

```python
#canny edge detection

canny = cv2.Canny(img,140,255,)
```

```python
#display images

titles = ['image', 'canny']
images = [img, canny]
```

```python
for i in range(2):
    plt.subplot(1,2,i+1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
    plt.savefig('output_edge.png', dpi=300)

plt.show()   
```

```python
#saving the images individually

cv2.imwrite('canny_edge_detection.tiff',canny)
```
